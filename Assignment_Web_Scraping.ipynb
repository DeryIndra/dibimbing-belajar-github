```python
import requests
import pandas as pd

url = 'https://carisayur.com/'
res = requests.get(url)
```

```python
res.content
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(res.content, 'html.parser')
```

```python
list_content_item = soup.select('.product-list')
```

```python
len(list_content_item)
```

```python
list_content_item[0]
```

```python
list_result_item = []
# for item in list_content_item[:1]:
for item in list_content_item:
  data = dict()

  print(item.select_one('a').get('href'))
  print(item.select_one('img').get('src'))
  #data collection
  data['sayur-item-link'] = item.select_one('a').get('href')
  data['sayur-img-src'] = item.select_one('img').get('src')

  list_result_item.append(data)
```

```python
pd.DataFrame(list_result_item)
```
